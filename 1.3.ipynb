```python
frame = pd.DataFrame(np.arange(9).reshape((3,3)), 
                     index=['2', '1','3'], 
                     columns=[ 'a', 'b', 'c'])
frame
```

```python
# 让行索引升序排序
frame.sort_index()
```

```python
# 让列索引升序排序
frame.sort_index(axis=1)
```

```python
# 让行索引降序排序
frame.sort_index(ascending=False)
```

```python
# 让列索引降序排序
frame.sort_index(axis=1, ascending=False)
```
