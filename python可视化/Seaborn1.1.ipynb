```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python

# 使用默认主题，如遇AttributeError请升级seaborn
#sns.set_theme()

# 加载seaborn的样例数据集
tips = pd.read_csv("tips.csv")

# 绘制图表
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", size="size")
```

```python
# 加载iris数据集
#iris = sns.load_dataset('iris', cache=True, data_home='searborn_sample')
iris = pd.read_csv('iris.csv')
sns.pairplot(iris,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 设置图表风格
plt.style.use(['tableau-colorblind10'])

# 字体问题，默认中文无法显示
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'

# 数据集
x = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
y = [123, 145, 152, 182, 147, 138, 189, 201, 203, 211, 201, 182]

# 设置画布大小
plt.figure(figsize=(8, 6))
# 画一个柱形图
sns.barplot(x=x, y=y)
# 设置坐标轴范围
plt.ylim(50, 250)
# 添加网格线
plt.grid(b=True, axis='y', linestyle='--', linewidth=1, color='grey')
# 添加标题
plt.title("销售趋势图", fontdict={'family':'Microsoft YaHei', 'size': 15}, loc='left')
# 添加xy轴标题
plt.xlabel("月份", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)
plt.ylabel("销售额（万元）", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)
```
