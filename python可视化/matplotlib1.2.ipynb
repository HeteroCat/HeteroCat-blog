```python
plt.figure(figsize=(8, 6))
x = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
y1 = [123, 145, 152, 182, 147, 138, 189, 201, 203, 211, 201, 182]
y2 = [102, 121, 138, 154, 171, 178, 199, 231, 228, 202, 231, 271]

plt.title("销售趋势图", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 15}, loc='left')
plt.plot(x, y1, linestyle='-.', marker='o', markersize=10, color='r', label='华东')  
plt.plot(x, y2, linestyle='-', marker='o', markersize=10, color='y', label='华中')  
plt.xlabel("月份", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)
plt.ylabel("销售额（万元）", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)
plt.legend(loc='best', fontsize=12) # best:matplotlib根据图表自动选择最优位置
```

```python
#作业
plt.figure(figsize=(12, 6))
plt.style.use(['seaborn'])
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'   #识别中文

cate = ['美国', '中国', '日本', '德国', '英国', '印度', '法国', '意大利', '加拿大', '韩国']  
GDP = [22.97, 17.73, 5.01, 4.20, 3.30, 3.08, 2.92, 2.12, 2.02, 1.80]
plt.barh(cate, GDP,color="#0aff00")  # 绘制图像
plt.xlabel("GDP/亿美元", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 12}, labelpad=10)
plt.title("2021年全球GDP排名", fontdict={'family':'Microsoft YaHei', 'color': 'k', 'size': 15}, loc='center')
for a, b in zip(GDP, cate):
    plt.text(a,b, str(a)+'亿', ha='left', fontdict={"size":14})
   
```
