```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
```

```python
fig = plt.figure(figsize=(9,6))
x = ["深圳", "广州", "北京", "上海"]
y = [1, 3, 2, 5]

plt.subplot(2,3, 1)
plt.bar(x, y)

plt.subplot(2, 3, 2)
plt.barh(x,y)

plt.subplot(2,3, 3)
plt.plot(x, y)

plt.subplot(2, 3, 4)
plt.stackplot(x, y)

plt.subplot(2,3, 5)
plt.pie(y, labels=x)

plt.subplot(2, 3, 6)
plt.scatter(x,y)

```
