```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
text = pd.read_csv(r'result.csv')
text.head()
```

```python
sex = text.groupby('Sex')['Survived'].sum()
sex.plot.bar()
plt.title('survived_count')
```

```python
a=text.groupby(['Sex','Survived'])['Survived'].count().unstack()
```

```python
die =a[0]
die.plot.bar()
```

```python
text.groupby(['Sex','Survived'])['Survived'].count().unstack().plot(kind='bar')
```

```python
text.groupby(['Sex','Survived'])['Survived'].count().unstack().plot(kind='bar',stacked='True')
```

```python
text.groupby(['Sex','Survived'])['Survived'].count().unstack().plot(kind='bar',stacked='True')
plt.title('survived_count')
plt.ylabel('count')
```

```python
fare = text.groupby(['Fare','Survived'])['Survived'].count().unstack()
fare  #0代表死，1代表活
```

```python
fare.plot()
```

```python
fare_sur1 = text.groupby(['Fare'])['Survived'].value_counts()
fare_sur1
```

```python
fig = plt.figure(figsize=(20, 18))
fare_sur1.plot(grid=True)
plt.legend()
plt.show()
```

```python
text.groupby(['Pclass','Survived'])['Survived'].count().unstack().plot(kind='bar')
```

```python
text.Age[text.Pclass == 1].plot(kind='kde')
text.Age[text.Pclass == 2].plot(kind='kde')
text.Age[text.Pclass == 3].plot(kind='kde')
plt.xlabel("age")
plt.legend((1,2,3),loc="best")
```

```python
text.Age[text.Survived ==0].hist(bins=5)
#不同年龄死亡情况
```

```python
text.Age[text.Survived ==1].hist(bins=5)
#不同年龄生存情况
```

```python
text.Age[text.Survived ==0].hist(bins=5,alpha=0.5)
text.Age[text.Survived ==1].hist(bins=5,alpha=0.5)
plt.legend((0,1))
plt.xlabel('age')
plt.ylabel('count')
```

```python
text.Age[text.Survived ==0].hist(bins=5,alpha=0.5,density=1)
text.Age[text.Survived ==1].hist(bins=5,alpha=0.5,density=1)
text.Age[text.Survived ==0].plot.density()
text.Age[text.Survived ==1].plot.density()
plt.legend((0,1))
plt.xlabel('age')
plt.ylabel('count')
```
