```python
import numpy as np
import pandas as pd
```

```python
text = pd.read_csv('result.csv')
```

```python
#GroupBy机制其实就是一个分类的过程，主要是分类后的运算会比较方便
```

```python
df  = text['Fare'].groupby(text['Sex'])
means = df.mean()
means
```

```python
df.describe()
```

```python
df = text['Age'].groupby(text['Sex'])
df.describe()
```

```python
survived_sex =text.groupby('Sex')['Survived'].sum()
```

```python
text.groupby('Pclass')['Survived'].sum()
```

```python
#思考心得 :从上面的数据分析我们可以看到女性的平均票价和存活人数都比男性要高，同时第二客仓等级的存活人数比较少。
```

```python
text.groupby('Sex').agg({'Fare': 'mean', 'Pclass': 'count'}).rename(columns=
                            {'Fare': 'mean_fare', 'Pclass': 'count_pclass'})
```

```python
text.groupby(['Pclass','Age'])['Fare'].mean().head()
```

```python
result = pd.merge(means,survived_sex,on='Sex')
result
```

```python
survived_age =text.groupby('Age')['Survived'].sum()
```

```python
#不同年龄的存活人数
survived_age
```

```python
max(survived_age)
```

```python
#找出最大值的年龄段
survived_age[survived_age.values==survived_age.max()]
```

```python
sum1 =text['Survived'].sum()
sum1
```

```python
rate= max(survived_age)/sum1
rate
```
